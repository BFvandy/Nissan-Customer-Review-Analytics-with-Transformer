# Pre-processing


```python
comments_2022 = pd.read_csv('SW_2022_verbatim_small_suv.csv', skiprows=12)
comments_2021 = pd.read_csv('SW_2021_verbatim_small_suv.csv', skiprows=12)
comments_2020 = pd.read_csv('SW_2020_verbatim_small_suv.csv', skiprows=12)
```

```python
comments_2022.insert(1, 'Study Year', '2022')
comments_2022.head(5)
```

```python
comments_2021.insert(1, 'Study Year', '2021')
comments_2021.head(5)
```

```python
comments_2020.insert(1, 'Study Year', '2020')
comments_2020.head(5)
```

```python
SW_dfs = [comments_2022, comments_2021, comments_2020]
```

```python
for df in SW_dfs:
    print(df.shape)
```

```python
all_SW_cmt = pd.concat(SW_dfs)
all_SW_cmt = all_SW_cmt.reset_index()
```

```python
# Tag Make and Model
Make = ['Nissan', 'Chevrolet', 'Ford', 'Ford', 'Honda', 'Hyundai', 'Kia', 'Mazda', 'Subaru', 'Toyota']
Model = ['Rogue', 'Equinox', 'Escape', 'Bronco Sport', 'CR-V', 'Tucson', 'Sportage', 'CX-5', 'Forester', 'RAV4']

def tag_model(brand):
  for tag in Model:
    if tag in brand:
      return tag
    else:
      pass

def tag_make(mak):
  for cm in Make:
    if cm in mak:
      return cm
    else:
      pass

mod = []
for i in range(len(all_SW_cmt['Purchased'])):
  brand = all_SW_cmt['Purchased'][i]
  mod.append(tag_model(brand))

all_SW_cmt.insert(0,'Model',mod)

car_make = []
for i in range(len(all_SW_cmt['Purchased'])):
  mak = all_SW_cmt['Purchased'][i]
  car_make.append(tag_make(mak))

all_SW_cmt.insert(0,'Make', car_make)
all_SW_cmt
```

```python
all_SW_cmt = all_SW_cmt.drop('index', axis=1)
```
